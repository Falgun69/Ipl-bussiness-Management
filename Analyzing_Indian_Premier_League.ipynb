```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample IPL Data
data =  pd.read_excel("Downloads/ipl_data.xlsx")
ipl_df = pd.DataFrame(data)

# 1. Basic Data Exploration
print(ipl_df.head())
print(ipl_df.info())
```

```python
# 2. Team Performance
team_wins = ipl_df["winner"].value_counts()
print("\n Team Wins:\n", team_wins)
```

```python
# 3. Plyer of the Match Analysis
player_of_match_counts = ipl_df["Player of the match"].value_counts()
print("\nPlayer of the Match Counts:\n", player_of_match_counts)
```

```python
# 4. Runs Scored Analysis
average_runs_per_match = ipl_df["run_scored"].mean()
print("\nAverage Runs Scored per Match:", average_runs_per_match)
```

```python
# 1. Bar Chart of Team Wins
plt.figure(figsize=(10, 6))
sns.countplot(x="winner", data=ipl_df)
plt.title("Number of Wins by Team")
plt.xlabel("Team")
plt.ylabel("Number of Wins")
plt.show()
```

```python
# 2. Histogram of Runs Scored
plt.figure(figsize=(8, 5))
plt.hist(ipl_df["run_scored"], bins=8)
plt.title("Distribution of Runs Scored per Match")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
```
